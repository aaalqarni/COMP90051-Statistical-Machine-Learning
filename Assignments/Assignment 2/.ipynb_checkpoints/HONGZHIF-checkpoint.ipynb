```python
# You can add additional imports here
import numpy as np
from matplotlib import pyplot as plt
import os
import random
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
from scipy.stats import entropy
from collections import defaultdict
```

## 0. Loading the dataset

```python
# do not edit this cell
# load the data files (download from the LMS)
embedded_images = np.load('images.npy')
labels = np.load('labels.npy')

# split into pool & testing
X_pool, X_test, y_pool, y_test = train_test_split(embedded_images, labels, 
                                                  test_size=0.5, random_state=1234, shuffle=True)

# sample a seed set
np.random.seed(1234)
label2id = defaultdict(list)
for i, label in enumerate(y_pool):
    label2id[label].append(i)
seed_set = []
for label, ids in label2id.items():
    seed_set.extend(np.random.choice(ids, size=10, replace=False))
```

## 1. Applying logistic regression

```python
def train_logistic_regression(X, y, **args):
    """
    Train a logistic regression model on dataset (X, y) and return trained model.
    X: matrix of real values, size n x d
    y: vector of string labels, size n
    args: optional arguments e.g., for hyper-parameters
    """
    # your code here
    # train model
    clf = LogisticRegression(C=args['c'], max_iter=args['max_iter'])
    clf.fit(X, y)
    return clf
```

```python
def evaluate_logistic_regression_accuracy(Xt, yt, model):
    """
    Apply logistic regression prediction on dataset Xt and evaluate accuracy against yt,
    returing the accuracy results as a scalar.
    Xt: matrix of real values, size m x d
    yt: vector of string labels, size m
    """
    # your code here
    
    # predict
    yp = model.predict(Xt)
    
    # evaluate model
    acc = accuracy_score(yt, yp)
    return acc
```

```python
# your code here for training, evaluating & plotting results

import warnings
warnings.filterwarnings('ignore')

all_C = [1e-3, 1e-2, 1e-1, 1, 10, 100, 1000]
all_acc = []
best_c = 0; best_model = None; best_acc = 0

for c in all_C:
    model = train_logistic_regression(X_pool, y_pool, c=c, max_iter=5000)
    acc = evaluate_logistic_regression_accuracy(X_test, y_test, model)
    
    if acc > best_acc:
        best_c = c; best_model = model; best_acc = acc
    
    all_acc.append(acc)
    print("c = {}, acc = {:.4f}".format(c, acc))
```

```python
# plot
plt.plot(np.log10(all_C), all_acc)
plt.xlabel('Logarithm of C')
plt.ylabel('Test Accuracy')
plt.title("Test Accuracy vs Logarithm of C")
plt.show()
```

## 2. Active learning framework with Random selection

```python
def random_select(X, model, **args):
    """
    Given an unlabelled dataset X, a matrix of n x d, and a model (not used)
    returns a vector of scores of length n. Each entry reflects the priority 
    of the corresponding instance. Higher means better.
    """
    # fill in
    scores = np.arange(X.shape[0], dtype=np.int)
    np.random.shuffle(scores)
    return scores
```

```python
def pool_based_active_learning(X_pool, y_pool, seed_ids,
                               train_func, select_func,
                               max_size, batch_size, **args):
    """
    Perform an active learning simulation, which starts by training on a seed set,
    then iteratively applies the selection function to rank instances in the pool,
    selects the top few instances which are included into the training set and the
    process repeats. 
        X_pool: matrix of n x d
        y_pool: vector of string labels, size n
        seed_ids: initial labelled set set, as a list of indices [0..n-1] into pool
        train_func: function which given (X, y, optional args) returns a trained model
        select_func: function which given (X, optional args) returns a sequence of scores
        max_size: stopping condition for active learning, when labelled data reaches given size
        batch_size: number of instances to be labelled in each iteration
        args: optional arguments passed to training and selection function
    returns the sequence of trained models 
    """
  
    # fill in
    seed_ids_ = seed_ids.copy()
    trained_models = []; indices = np.array([i for i in range(len(X_pool))])
    
    i = 0
    while len(seed_ids_) < max_size:
        # reveal seed dataset labels
        X_seed, y_seed = X_pool[seed_ids_], y_pool[seed_ids_]
        
        # train a model
        model = train_func(X_seed, y_seed, c=args['c'], max_iter=args['max_iter'])
        
        # compute scores
        scores = select_func(X_pool[indices], model)
        
        # select next batch of seed ids
        top_batch_idx = np.argsort(-scores)[:batch_size]  # descending order
        
        # add new seed ids into `seed_ids`
        seed_ids_.extend(top_batch_idx.tolist())
        
        # remove new seed ids from `indices`
        indices = np.delete(indices, top_batch_idx)
        
        # save current model
        trained_models.append(model)
        i += 1
        print('Saved %d models.' % i if i > 1 else "Saved 1 model.")
        
    return trained_models
    
```

```python
batch = 60
max_size = 3000
models_random = pool_based_active_learning(X_pool, y_pool, seed_set, 
                                    train_logistic_regression, random_select, 
                                    max_size, batch, c=0.1, max_iter=5000)
```

```python
## your code here for evaluation of accuracy and plotting of results

# evaluation
iterations = []
all_test_acc = []

for i, model in enumerate(models_random):
    test_acc = evaluate_logistic_regression_accuracy(X_test, y_test, model)
    iterations.append(i+1)
    all_test_acc.append(test_acc)
    
    print('Model %d with test accuracy %.4f' % (i+1, test_acc))
```

```python
# plot
plt.plot(iterations, all_test_acc)
plt.xlabel('Iteration')
plt.ylabel('Test Accuracy')
plt.title("Test Accuracy vs Iterations on Random Sampling")
plt.show()
```

## 3. Uncertainty sampling

```python
def logistic_regression_entropy_select(X, model, **args):
    """
    Given an unlabelled dataset X, a matrix of n x d, and a discriminative model 
    P(y|x), returns a vector of n entropy values.
    """
    # fill in
    prob = model.predict_proba(X)
    scores = entropy(prob, base=2, axis=1)
    return scores
    
```

```python
models_us = pool_based_active_learning(X_pool, y_pool, seed_set, 
                                 train_logistic_regression, 
                                 logistic_regression_entropy_select, 
                                 max_size, batch, c=0.1, max_iter=5000)
```

## 4. Query by committee

```python
# don't forget to provide function descriptive comments, like those provided in templates above

def query_by_committee_vote_entropy(X, models, **args):
    vote = []
    for model in models:
        pred_labels = np.argmax(model.predict_proba(X), axis=1)  # predict labels
        vote.append(pred_labels)  # append to a list
    vote = np.stack(vote, axis=1)  # concatenation
    p_vote = np.stack([np.bincount(vote[i], minlength=30) / 30 for i in range(vote.shape[0])], axis=0)  # compute probability
    
    scores = entropy(p_vote, axis=1)  # compute entropy
    return scores

def query_by_committee_soft_vote_entropy(X, models, **args):
    vote = []
    for model in models:
        probs = model.predict_proba(X)  # predict probability
        vote.append(probs)  # append results to a list
    vote_prob = np.stack(vote)  # concatenation
    soft_vote_prob = np.mean(vote_prob, axis=0)  # compute means for all instances
    
    scores = entropy(soft_vote_prob, axis=1)  # compute entropy
    return scores

def query_by_committee_KL(X, models, **args):
    vote = []
    for model in models:
        probs = model.predict_proba(X)  # predict probability
        vote.append(probs)  # append results to a list
    vote_prob = np.stack(vote)  # concatenation
    soft_vote_prob = np.mean(vote_prob, axis=0)  # compute means for all instances
    
    # compute KL divergence for each committee and for each instances
    n_committee, n_instances, _ = vote_prob.shape
    KL_divergence = np.empty([n_instances, n_committee])
    
    for i in range(n_instances):
        for j in range(n_committee):
            KL_divergence[i, j] = entropy(soft_vote_prob[i], qk=vote_prob[j, i])
    
    scores = np.max(KL_divergence, axis=1)  # compute final score
    return scores
    
```

```python
def train_committee(X_pool, y_pool, **args):
    seed_ids = args['seed_ids']
    train_func = args['train_func']
    select_func = args['select_func']
    max_size = args['max_size']
    batch_size = args['batch_size']
    
    seed_ids_ = seed_ids.copy()
    trained_models_1 = []; trained_models_2 = []
    indices = np.array([i for i in range(len(X_pool))])
    
    i = 0
    while len(seed_ids_) < max_size:
        # reveal seed dataset labels
        X_seed, y_seed = X_pool[seed_ids_], y_pool[seed_ids_]
        
        # bagging sampling
        seed_indices = np.arange(X_seed.shape[0])
        
        # shuffle seed indices
        sample_indices = np.random.choice(seed_indices, seed_indices.shape[0], True)
        X_seed_1, y_seed_1 = X_seed[sample_indices], y_seed[sample_indices]
        
        # shuffle again
        sample_indices = np.random.choice(seed_indices, seed_indices.shape[0], True)
        X_seed_2, y_seed_2 = X_seed[sample_indices], y_seed[sample_indices]
        
        # train a model
        model_1 = train_func(X_seed_1, y_seed_1, c=args['c'], max_iter=args['max_iter'])
        model_2 = train_func(X_seed_2, y_seed_2, c=args['c'], max_iter=args['max_iter'])
        
        # compute scores
        scores = select_func(X_pool[indices], [model_1, model_2])
        
        # select next batch of seed ids
        top_batch_idx = np.argsort(-scores)[:batch_size]  # descending order
        
        # add new seed ids into `seed_ids`
        seed_ids_.extend(top_batch_idx.tolist())
        
        # remove new seed ids from `indices`
        indices = np.delete(indices, top_batch_idx)
        
        # save current model
        trained_models_1.append(model_1)
        trained_models_2.append(model_2)
        
        print("Model 1 accuracy: %.4f" % (evaluate_logistic_regression_accuracy(X_seed_1, y_seed_1, model_1)))
        print("Model 2 accuracy: %.4f" % (evaluate_logistic_regression_accuracy(X_seed_2, y_seed_2, model_2)))
        
        i += 1
        print('Saved %d models.' % i if i > 1 else "Saved 1 model.")
        
    return trained_models_1, trained_models_2
```

```python
# vote entropy

# train model
models_ve = train_committee(X_pool, y_pool, seed_ids=seed_set, 
                            train_func=train_logistic_regression, 
                            select_func=query_by_committee_vote_entropy, 
                            max_size=max_size, batch_size=batch,
                            c=0.1, max_iter=5000)
```

```python
# evaluation
trained_models_1, trained_models_2 = models_ve

iterations = []
all_test_acc_1 = []
all_test_acc_2 = []

for i, model in enumerate(trained_models_1):
    test_acc = evaluate_logistic_regression_accuracy(X_test, y_test, model)
    iterations.append(i+1)
    all_test_acc_1.append(test_acc)
    
    print('Model 1-%d with test accuracy %.4f' % (i+1, test_acc))

for i, model in enumerate(trained_models_2):
    test_acc = evaluate_logistic_regression_accuracy(X_test, y_test, model)
    all_test_acc_2.append(test_acc)
    
    print('Model 2-%d with test accuracy %.4f' % (i+1, test_acc))
```

```python
# plot
plt.plot(iterations, all_test_acc_1, label='ensemble 1')
plt.plot(iterations, all_test_acc_2, label='ensemble 2')
plt.legend(loc='upper left')
plt.xlabel('Iteration')
plt.ylabel('Test Accuracy')
plt.title("Query by Committee Vote Entropy")
plt.show()
```
